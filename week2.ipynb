# Folium 사용하기

```python
import folium

x = df_seoul_hospital["위도"].mean()
y = df_seoul_hospital["경도"].mean()
```

```python
df_seoul_hospital.tail(1)
```

```python
map = folium.Map(location = [x, y], zoom_start=12)

for n in df_seoul_hospital.index:
    name = df_seoul_hospital.loc[n,"상호명"]
    addr = df_seoul_hospital.loc[n,"도로명주소"]
    popup = f"{name}-{addr}"
    location = [df_seoul_hospital.loc[n,"위도"],df_seoul_hospital.loc[n,"경도"]]
    folium.Marker(
        location = location, 
        popup = popup
    ).add_to(map)
map
```
